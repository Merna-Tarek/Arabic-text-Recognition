```python
from IPython.display import display, Image
import matplotlib.pyplot as plt
import pytesseract
from gtts import gTTS
#to play the voice
import IPython.display as ipd
from requests import get
```

```python
def download(url, file_name):
    # open in binary mode
    with open(file_name, "wb") as file:
        # get request
        response = get(url)
        # write to file
        file.write(response.content)

download("https://github.com/tesseract-ocr/tessdata/raw/master/ara.traineddata","ara.traineddata")
```

```python
import cv2

filename="Arabic_OCR.png" 
img_cv = cv2.imread(filename)
plt.figure(figsize=(20, 20))
plt.imshow(img_cv,cmap = 'gray')
plt.title('template'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'  
arabic_text=pytesseract.image_to_string(img_cv , lang='ara',config= ".")

print(arabic_text)
```

```python
tts = gTTS(arabic_text, lang='ar')
tts.save('test.mp3')
```

```python
audio_path="test.mp3"

ipd.Audio(audio_path, autoplay=True)
```
